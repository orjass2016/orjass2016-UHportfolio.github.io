```python
# ============================================================
# 1. Reproducibility helper
# ============================================================

def set_seed(seed: int = 42) -> None:
    """
    Make training as reproducible as possible.
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
```

```python
# ============================================================
# 2. Dataset loading & transforms
# ============================================================

def load_pickle_dataset(path: str) -> Tuple[np.ndarray, np.ndarray]:
    """
    Flexible loader for train-70_.pkl / validation-10_.pkl files.

    Assumes the pickle contains either:
      - dict with ("images", "labels") or ("data", "targets"), OR
      - tuple/list like (X, y)

    Returns:
      X: numpy array of shape (N, H, W, C) or (N, C, H, W)
      y: numpy array of shape (N,)
    """
    with open(path, "rb") as f:
        obj = pickle.load(f)

    if isinstance(obj, dict):
        if "images" in obj and "labels" in obj:
            X, y = obj["images"], obj["labels"]
        elif "data" in obj and "targets" in obj:
            X, y = obj["data"], obj["targets"]
        else:
            values = list(obj.values())
            if len(values) < 2:
                raise ValueError(
                    "Dict pickle does not contain enough arrays (need images + labels)."
                )
            X, y = values[0], values[1]
    elif isinstance(obj, (list, tuple)) and len(obj) >= 2:
        X, y = obj[0], obj[1]
    else:
        raise ValueError("Unsupported pickle structure; expected dict or (X, y).")

    X = np.array(X)
    y = np.array(y)
    return X, y


class TinyImageNetDataset(Dataset):
    """
    Wraps in-memory numpy arrays into a PyTorch Dataset.
    Works for both labeled (train/val) and unlabeled (test) data.
    """

    def __init__(self, images: np.ndarray, labels: np.ndarray = None, transform=None):
        """
        images: numpy array of:
          - (N, H, W, C) [channel-last] OR
          - (N, C, H, W) [channel-first]
        labels: numpy array of shape (N,) with class indices (0..14) or None
        """
        assert images.ndim == 4, "Expected images with shape (N, H, W, C) or (N, C, H, W)"
        self.images = images
        self.labels = labels
        self.transform = transform
        self.has_labels = labels is not None

    def __len__(self) -> int:
        return self.images.shape[0]

    def _to_pil(self, img: np.ndarray) -> Image.Image:
        """
        Ensure image is HWC uint8 and convert to PIL.
        """
        if img.ndim != 3:
            raise ValueError("Each image must be 3D (H, W, C) or (C, H, W).")

        # If channel-first (C, H, W), transpose to (H, W, C)
        if img.shape[0] == 3 and img.shape[-1] != 3:
            img = np.transpose(img, (1, 2, 0))

        # Clip & convert to uint8 in case it's floats
        if img.dtype != np.uint8:
            img = np.clip(img, 0, 255).astype(np.uint8)

        return Image.fromarray(img)

    def __getitem__(self, idx: int):
        img = self.images[idx]
        img = self._to_pil(img)

        if self.transform is not None:
            img = self.transform(img)

        if self.has_labels:
            label = int(self.labels[idx])
            return img, label
        else:
            return img


def get_transforms() -> Tuple[transforms.Compose, transforms.Compose]:
    """
    Train & validation/test transforms.
    - Data augmentation only on training set.
    - Normalize with ImageNet-style mean/std.
    """
    mean = [0.485, 0.456, 0.406]
    std = [0.229, 0.224, 0.225]

    train_tfms = transforms.Compose([
        transforms.RandomResizedCrop(64, scale=(0.8, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(10),
        transforms.ColorJitter(
            brightness=0.1, contrast=0.1, saturation=0.1, hue=0.02
        ),
        transforms.ToTensor(),
        transforms.Normalize(mean, std),
    ])

    val_tfms = transforms.Compose([
        transforms.Resize(64),
        transforms.CenterCrop(64),
        transforms.ToTensor(),
        transforms.Normalize(mean, std),
    ])

    return train_tfms, val_tfms
```

```python
# ============================================================
# 3. CNN model architecture (Sequential CNN, optimized head)
# ============================================================

class TinyImageNetCNN(nn.Module):
    """
    Custom Sequential CNN:
    - Only uses Conv2D, Pooling, BatchNorm, Dropout, Flatten, Dense (Linear)
    - No pretrained features, no residual/skip connections, no attention.
    """

    def __init__(self, num_classes: int = 15):
        super().__init__()

        # Feature extractor (4 conv blocks)
        self.features = nn.Sequential(
            # Block 1: 3x64x64 -> 64x32x32
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),

            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),

            nn.MaxPool2d(kernel_size=2, stride=2),  # 64x32x32
            nn.Dropout(0.2),

            # Block 2: 64x32x32 -> 128x16x16
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),

            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),

            nn.MaxPool2d(kernel_size=2, stride=2),  # 128x16x16
            nn.Dropout(0.3),

            # Block 3: 128x16x16 -> 256x8x8
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.Conv2d(256, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.MaxPool2d(kernel_size=2, stride=2),  # 256x8x8
            nn.Dropout(0.4),

            # Block 4: 256x8x8 -> 512x4x4 -> GAP 512x1x1
            nn.Conv2d(256, 512, kernel_size=3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),

            nn.MaxPool2d(kernel_size=2, stride=2),    # 512x4x4
            nn.AdaptiveAvgPool2d((1, 1)),             # 512x1x1 (global average pooling)
        )

        # Classifier head (much lighter than original 8192→512)
        self.classifier = nn.Sequential(
            nn.Flatten(),                  # 512 * 1 * 1 = 512
            nn.Dropout(0.5),
            nn.Linear(512, 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x = self.features(x)
        x = self.classifier(x)
        return x
```

```python
# ============================================================
# 4. Training and evaluation loops
# ============================================================

def train_model(
    train_loader: DataLoader,
    val_loader: DataLoader,
    num_classes: int = 15,
    epochs: int = 40,
    patience: int = 7,
    lr: float = 1e-3,
    weight_decay: float = 1e-4,
    device: torch.device = None,
    save_path: str = "model.pth",
) -> Tuple[TinyImageNetCNN, Dict[str, List[float]]]:
    """
    Train the CNN with early stopping, saving the best weights to model.pth.

    Returns:
      model: best model (with best validation accuracy loaded)
      history: dict with loss/accuracy curves
    """
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model = TinyImageNetCNN(num_classes=num_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(
        model.parameters(), lr=lr, weight_decay=weight_decay
    )
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode="max", factor=0.5, patience=3,
    )

    history: Dict[str, List[float]] = {
        "train_loss": [],
        "val_loss": [],
        "train_acc": [],
        "val_acc": [],
    }

    best_val_acc = 0.0
    best_state_dict = None
    epochs_no_improve = 0

    for epoch in range(epochs):
        # ---- Train ----
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images = images.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * labels.size(0)
            _, preds = torch.max(outputs, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        train_loss = running_loss / total
        train_acc = correct / total

        # ---- Validation ----
        model.eval()
        val_loss_sum = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for images, labels in val_loader:
                images = images.to(device, non_blocking=True)
                labels = labels.to(device, non_blocking=True)

                outputs = model(images)
                loss = criterion(outputs, labels)

                val_loss_sum += loss.item() * labels.size(0)
                _, preds = torch.max(outputs, dim=1)
                val_correct += (preds == labels).sum().item()
                val_total += labels.size(0)

        val_loss = val_loss_sum / val_total
        val_acc = val_correct / val_total

        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)

        scheduler.step(val_acc)

        print(
            f"Epoch [{epoch+1}/{epochs}]  "
            f"Train Loss: {train_loss:.4f}  Train Acc: {train_acc:.4f}  "
            f"Val Loss: {val_loss:.4f}  Val Acc: {val_acc:.4f}"
        )

        # ---- Early stopping ----
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_state_dict = model.state_dict()
            epochs_no_improve = 0
            torch.save(best_state_dict, save_path)
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print("Early stopping triggered.")
                break

    if best_state_dict is not None:
        model.load_state_dict(best_state_dict)

    return model, history
```

```python
# ============================================================
# 5. Load & predict for instructor’s hidden test set
# ============================================================

def load_trained_model(
    weights_path: str = "model.pth",
    num_classes: int = 15,
    device: torch.device = None,
) -> TinyImageNetCNN:
    """
    Load model architecture + weights.

    This is what the instructor will call by design:
      1. Instantiate TinyImageNetCNN
      2. Load weights from model.pth
      3. Return model in eval mode
    """
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model = TinyImageNetCNN(num_classes=num_classes)
    state_dict = torch.load(weights_path, map_location=device)
    model.load_state_dict(state_dict)
    model.to(device)
    model.eval()
    return model


def predict(
    model: TinyImageNetCNN,
    test_images: np.ndarray,
    batch_size: int = 64,
    device: torch.device = None,
) -> np.ndarray:
    """
    Predict class indices for a batch of test images.

    Args:
      model: loaded TinyImageNetCNN in eval() mode
      test_images: numpy array of shape (N, H, W, C) or (N, C, H, W)
      batch_size: inference batch size
      device: torch.device (if None, auto-detect)

    Returns:
      preds: numpy array of shape (N,) with predicted class indices.
    """
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    _, val_tfms = get_transforms()
    test_dataset = TinyImageNetDataset(test_images, labels=None, transform=val_tfms)
    test_loader = DataLoader(
        test_dataset, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True
    )

    all_preds: List[np.ndarray] = []

    model.eval()
    with torch.no_grad():
        for images in test_loader:
            images = images.to(device, non_blocking=True)
            outputs = model(images)
            preds = torch.argmax(outputs, dim=1)
            all_preds.append(preds.cpu().numpy())

    all_preds = np.concatenate(all_preds, axis=0)
    return all_preds
```

```python
# ============================================================
# 6. Visualization helpers (optional, not required by rubric)
# ============================================================

def plot_confusion_heatmap(
    y_true: np.ndarray,
    y_pred: np.ndarray,
    class_names: List[str] | None = None,
    normalize: bool = True,
    figsize: Tuple[int, int] = (8, 6),
):
    """
    Heatmap of confusion matrix (True vs Predicted).

    - Picks the actually used labels from y_true/y_pred.
    - If class_names is provided but length doesn't match, falls back to numeric labels.
    """
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    # Only the labels that actually appear
    labels = np.unique(np.concatenate([y_true, y_pred]))
    cm = confusion_matrix(y_true, y_pred, labels=labels)

    if normalize:
        cm = cm.astype(np.float32)
        row_sums = cm.sum(axis=1, keepdims=True) + 1e-8
        cm = cm / row_sums

    # Decide what to show as tick labels
    if class_names is None or len(class_names) != len(labels):
        display_names = [str(l) for l in labels]
    else:
        display_names = class_names

    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(cm, interpolation="nearest", cmap="viridis")
    ax.figure.colorbar(im, ax=ax)

    ax.set(
        xticks=np.arange(len(display_names)),
        yticks=np.arange(len(display_names)),
        xticklabels=display_names,
        yticklabels=display_names,
        xlabel="Predicted Label",
        ylabel="True Label",
        title="Confusion Matrix Heatmap",
    )

    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    fmt = ".2f" if normalize else "d"
    thresh = cm.max() / 2.0
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(
                j,
                i,
                format(cm[i, j], fmt),
                ha="center",
                va="center",
                color="white" if cm[i, j] > thresh else "black",
            )

    fig.tight_layout()
    plt.show()


def get_logits_and_labels(
    model: TinyImageNetCNN,
    data_loader: DataLoader,
    device: torch.device = None,
) -> Tuple[np.ndarray, np.ndarray]:
    """
    Run the model over a DataLoader and gather logits + labels.
    Useful for histograms and boxplots of predicted probabilities.
    """
    if device is None:
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model.eval()
    all_logits: List[torch.Tensor] = []
    all_labels: List[torch.Tensor] = []

    with torch.no_grad():
        for images, labels in data_loader:
            images = images.to(device, non_blocking=True)
            outputs = model(images)   # (B, num_classes)
            all_logits.append(outputs.cpu())
            all_labels.append(labels)

    all_logits_tensor = torch.cat(all_logits, dim=0)
    all_labels_tensor = torch.cat(all_labels, dim=0)

    return all_logits_tensor.numpy(), all_labels_tensor.numpy()


def plot_probability_histogram(
    logits: np.ndarray,
    class_idx: int,
    bins: int = 20,
):
    """
    Histogram of predicted probabilities for a chosen class index.
    """
    logits_tensor = torch.tensor(logits)
    probs = torch.softmax(logits_tensor, dim=1).numpy()
    class_probs = probs[:, class_idx]

    plt.figure(figsize=(6, 4))
    plt.hist(class_probs, bins=bins, edgecolor="black")
    plt.xlabel(f"Predicted probability for class {class_idx}")
    plt.ylabel("Count")
    plt.title("Histogram of predicted probabilities")
    plt.tight_layout()
    plt.show()


def plot_classwise_boxplot(
    logits: np.ndarray,
    labels: np.ndarray,
    class_names: List[str],
):
    """
    Boxplot of correct-class probabilities per class.
    Shows how confident the model is on each class.
    """
    logits_tensor = torch.tensor(logits)
    probs = torch.softmax(logits_tensor, dim=1).numpy()
    labels = np.array(labels)

    data = []
    for c in range(len(class_names)):
        mask = labels == c
        if np.sum(mask) == 0:
            data.append([0.0])
        else:
            data.append(probs[mask, c])

    plt.figure(figsize=(10, 5))
    plt.boxplot(data, labels=class_names, showmeans=True)
    plt.xticks(rotation=45, ha="right")
    plt.ylabel("Correct-class probability")
    plt.title("Class-wise Probability Distribution (Boxplot)")
    plt.tight_layout()
    plt.show()


def plot_training_curves(history: Dict[str, List[float]]):
    """
    Simple line plots for training/validation loss & accuracy.
    """
    epochs = range(1, len(history["train_loss"]) + 1)

    plt.figure(figsize=(12, 4))
    # Loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, history["train_loss"], label="Train Loss")
    plt.plot(epochs, history["val_loss"], label="Val Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training vs Validation Loss")
    plt.legend()

    # Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history["train_acc"], label="Train Acc")
    plt.plot(epochs, history["val_acc"], label="Val Acc")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Training vs Validation Accuracy")
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
# ============================================================
# 7. Example main() for your own training runs
# ============================================================

def main():
    """
    Example usage (for your own training and validation only):

      python groupX_cnn.py

    Trains on
      - train-70_.pkl (70%)
      - validation-10_.pkl (10%)

    The hidden 20% test set is only used by the instructor.
    """
    set_seed(42)

    train_path = "train-70_.pkl"
    val_path   = "validation-10_.pkl"

    print("Loading data...")
    X_train, y_train = load_pickle_dataset(train_path)
    X_val,   y_val   = load_pickle_dataset(val_path)

    # ----- figure out how many classes we actually have -----
    # (avoids "Target XX is out of bounds" errors)
    max_label = int(max(y_train.max(), y_val.max()))
    num_classes = max_label + 1
    print(f"Detected label range: 0 .. {max_label}  -> num_classes = {num_classes}")

    train_tfms, val_tfms = get_transforms()

    train_ds = TinyImageNetDataset(X_train, y_train, transform=train_tfms)
    val_ds   = TinyImageNetDataset(X_val,   y_val,   transform=val_tfms)

    # Adjust num_workers according to your OS / RAM
    train_loader = DataLoader(
        train_ds, batch_size=64, shuffle=True,  num_workers=2, pin_memory=True
    )
    val_loader = DataLoader(
        val_ds,   batch_size=64, shuffle=False, num_workers=2, pin_memory=True
    )

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Training on device: {device}")

    model, history = train_model(
        train_loader,
        val_loader,
        num_classes=num_classes,   # <--- use auto-detected class count
        epochs=40,
        patience=8,
        lr=1e-3,
        weight_decay=1e-4,
        device=device,
        save_path="model.pth",
    )

    # Optional: visualize curves
    plot_training_curves(history)

    # Example: confusion matrix heatmap on validation
    model.eval()
    all_val_labels = []
    all_val_preds  = []
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device, non_blocking=True)
            outputs = model(images)
            preds = torch.argmax(outputs, dim=1)
            all_val_labels.append(labels.numpy())
            all_val_preds.append(preds.cpu().numpy())

    all_val_labels = np.concatenate(all_val_labels)
    all_val_preds  = np.concatenate(all_val_preds)

    # build class_names matching num_classes so the heatmap shape lines up
    class_names = [f"class_{i}" for i in range(num_classes)]
    plot_confusion_heatmap(all_val_labels, all_val_preds)


if __name__ == "__main__":
    main()
```
